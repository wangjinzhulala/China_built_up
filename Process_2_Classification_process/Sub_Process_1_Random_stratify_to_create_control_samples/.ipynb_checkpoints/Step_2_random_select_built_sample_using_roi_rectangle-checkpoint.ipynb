```python
import ee
import subprocess
import pprint
import pandas as pd
```

```python
ee.Initialize()
```

##### get the points from gee_asset

```python
# get the roi-points
points_str = subprocess.run('earthengine ls users/wangjinzhulala/China_Database',shell=True, capture_output=True).stdout
```

```python
# format and transform the str to asset_path
asset_path = [s for s in points_str.decode('UTF-8').split('\r\n') if 'Points' in s]
```

##### merge all points to one

```python
merged_pt = ee.FeatureCollection(asset_path[0])

for i in range(1,len(asset_path)):
    
    pt = ee.FeatureCollection(asset_path[i])
    merged_pt = merged_pt.merge(pt)
```

##### join the roi_rectangle to roi_points

```python
roi_rect = ee.FeatureCollection("users/wangjinzhulala/China_Database/China_rect")
```

```python
# construct a spatial filter
spatialFilter = ee.Filter.intersects(leftField ='.geo',
                                     rightField = '.geo',
                                     maxError = 10)

# construct a save all join.
saveAllJoin = ee.Join.saveAll(matchesKey = 'points')

# apply the join
intersect = saveAllJoin.apply(roi_rect, merged_pt, spatialFilter)
```

##### randome select one feature from the intersect

```python
# 1) get the points that intersected with each rectangle
# 2) random the points and select one out of them
intersect_random = intersect.map(lambda x: ee.FeatureCollection(ee.List(x.get('points')))
                                              .randomColumn(seed=101)
                                              .sort('random')
                                              .first())
```

```python
# export the points
task = ee.batch.Export.table.toAsset(collection   = intersect_random.select('coordinates'),
                                     description = 'intersect_random_points',
                                     assetId     = 'users/wangjinzhulala/China_Database/intersect_random_points')

task.start()
```
