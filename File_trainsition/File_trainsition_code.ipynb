```python
import ee
```

```python
ee.Initialize()
```

##### Define the files from ee_path

```python
# fetch the Fourier imgs 
Path_huadong = 'users/Built_up_china/Built_up_mapping/01_Data_preparation/01_Fourier_img'

Fourier_huadong =  [ee.Image(f'{Path_huadong}/Fourier_1990_1992')]
```

##### Export the img

```python
name = 'Fourier_1990_1992_huadong'

# imprt North_China_Plain boundary
Target_area = ee.FeatureCollection("users/wangjinzhulala/China_built_up/01_Boundary_shp/China_zone")\
                .filterMetadata('NAME1','equals','华东')

# crate the export task
task = ee.batch.Export.image.toCloudStorage(
    image = Fourier_huadong[0],
    description=name,
    bucket='north_china_plain',
    fileNamePrefix=name,
    region=Target_area.geometry(),
    scale=30,
    maxPixels=int(1e13),
    skipEmptyTiles=True)

task.start()

# print out the process
print(f'Export {name}!')
```
