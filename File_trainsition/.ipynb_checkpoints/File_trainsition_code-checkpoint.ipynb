```python
import ee
```

```python
ee.Initialize()
```

#####  Construct the year_ranges

```python
# Define the time-range
# just get one_range from every different sensors
year_range = [(f'{i}_{i+2}') for i in range(1990,2020,3)][:1]

year_range
```

```python
Target_area_cn = '华东'
Target_area_en = 'huadong'
```

##### Define the files from ee_path

```python
# fetch the Fourier imgs 
Fourier_path_huadong = 'users/Built_up_china/Built_up_mapping/01_Data_preparation/01_Fourier_img/Fourier'
Fourier_huadong =  [ee.Image(f'{Fourier_path_huadong}_{year}').setMulti(['year',year,'Type','Fourier']) for year in year_range]
```

```python
# fetch the Landsat imgs
Landsat_path_huadong = 'users/Built_up_china/Built_up_mapping/01_Data_preparation/02_Landsat_cloud_free_img/Landsat_cloud_free'
Landsat_huadong =  [ee.Image(f'{Landsat_path_huadong}_{year}').setMulti(['year',year,'Type','Landsat'])  for year in year_range]
```

```python
# combine all img together
All_imgs  = Fourier_huadong + Landsat_huadong
```

##### Export the img

```python
for img in All_imgs:
    
    # cast img to ee.Image
    img = ee.Image(img)
    
    # get the metedata of the img
    year = img.get('year').getInfo()
    Type = img.get('Type').getInfo()
    
    # construct the name for export
    name = f'{Type}_{year}_{Target_area_en}'

       
    # imprt North_China_Plain boundary
    Target_area = ee.FeatureCollection("users/wangjinzhulala/China_built_up/01_Boundary_shp/China_zone")\
                    .filterMetadata('NAME1','equals',Target_area_cn)

    # crate the export task
    task = ee.batch.Export.image.toCloudStorage(
        image = img,
        description=name,
        bucket='north_china_plain',
        fileNamePrefix=name,
        region=Target_area.geometry(),
        scale=30,
        maxPixels=int(1e13),
        skipEmptyTiles=True)

    task.start()

    # print out the process
    print(f'Export {name}!')
```
