```python
import subprocess
```

```python
# need install Google-SDK first
imgs = subprocess.run('gsutil ls gs://north_china_plain/Base_images', shell=True, check=True,capture_output=True).stdout
```

```python
# conver byte to string
img_path = [s for s in imgs.decode('UTF-8').split('\r\n') if 'tif' in s]
```

```python
# get the url
img_urls = [s.replace('gs://','https://storage.googleapis.com/') for s in img_path ]
```

```python
# wirte urls to txt
with open('GS_download_urls.txt','w') as f:
    for url in img_urls:
        f.write(f'{url}\n')
```
